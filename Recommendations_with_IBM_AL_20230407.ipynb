```python
# create the user-article matrix with 1's and 0's

def create_user_item_matrix(df):
    '''
    INPUT:
    df - pandas dataframe with article_id, title, user_id columns
    
    OUTPUT:
    user_item - user item matrix 
    
    Description:
    Return a matrix with user ids as rows and article ids on the columns with 1 values where a user interacted with 
    an article and a 0 otherwise
    '''
    # Fill in the function here
    df_clean = df.drop_duplicates(subset=["article_id","user_id"])
    
    # Piovt article_ids to column headers
    matrix_w_nulls = df_clean.groupby(["user_id","article_id"])["title"].count().unstack()
    
    # Zero fill for null values
    user_item = matrix_w_nulls.fillna(0)
    
    
    return user_item #user_item # return the user_item matrix 

user_item = create_user_item_matrix(df)
```

```python
def find_similar_users(user_id, user_item=user_item):
    '''
    INPUT:
    user_id - (int) a user_id
    user_item - (pandas dataframe) matrix of users by articles: 
                1's when a user has interacted with an article, 0 otherwise
    
    OUTPUT:
    similar_users - (list) an ordered list where the closest users (largest dot product users)
                    are listed first
    
    Description:
    Computes the similarity of every pair of users based on the dot product
    Returns an ordered
    
    '''
#This commented out portion is the old method before re-factoring. It adds extra steps to convert dataframes to np.arrays 
#and retreive the specific user_id #array first. This causes a slow down and increase in complexity. To address the
#required feedback, I used an alternative approach of computing the similarity of each user vs every user in the 
#user_item matrix before finding the specific user from the index. I also used the dataframes directly and the dot
#and transpose methods.

#     # get row from user_item matrix for the user_id provided
#     user_articles = np.array(user_item.loc[user_id])
    
#     # compute similarity of each user to the provided user
#     dotp = np.dot(user_articles, np.array(user_item).T)
    
#     # convert results to Series with and index of user_ids
#     simiarlity = pd.Series(dotp, index=user_item.index)

#     # sort by similarity
#     simiarlity.sort_values(ascending=False, inplace=True)
    
#     # remove the own user's id 
#     simiarlity.drop(index=user_id, inplace=True)

#     # create list of just the ids
#     most_similar_users = simiarlity.index.tolist()
    
    
    # compute similarity of each user vs every user in the user_item matrix
    users_similar = user_item.dot(user_item.T)
    
    # retrieve sorted similarity scores to each user for the provided user_id
    user_id_sorted = users_similar[user_id].sort_values(ascending=False)

    # remove the own user's id 
    user_id_sorted.drop(index=user_id, inplace=True)

    # create list of just the ids
    most_similar_users = user_id_sorted.index.tolist()
    
    
       
    return most_similar_users # return a list of the users in order from most to least similar
        
```

```python
def get_article_names(article_ids, df=df):
    '''
    INPUT:
    article_ids - (list) a list of article ids
    df - (pandas dataframe) df as defined at the top of the notebook
    
    OUTPUT:
    article_names - (list) a list of article names associated with the list of article ids 
                    (this is identified by the title column)
    '''
    # Your code here
    
    # store list of input ids for lookup
    ids = article_ids
    
    # create lookup table
    id_title_ref = pd.DataFrame(df.groupby(['article_id','title']).groups.keys() \
                             ).rename(columns={0:'article_id',1:'title'})
    
    # convert the ids column to string to ensure matching on same data type
    id_title_ref['article_id'] = id_title_ref['article_id'].astype(str)
    
    #retrieve article titles for input ids
    article_names = id_title_ref[id_title_ref['article_id'].isin(ids)]['title'].tolist()
    
        
    return article_names # Return the article names associated with list of article ids
```

```python
def get_user_articles(user_id, user_item=user_item):
    '''
    INPUT:
    user_id - (int) a user id
    user_item - (pandas dataframe) matrix of users by articles: 
                1's when a user has interacted with an article, 0 otherwise
    
    OUTPUT:
    article_ids - (list) a list of the article ids seen by the user
    article_names - (list) a list of article names associated with the list of article ids 
                    (this is identified by the doc_full_name column in df_content)
    
    Description:
    Provides a list of the article_ids and article titles that have been seen by a user
    '''
    # Your code here
    
    #retrieve article indexes a specified user_id interacted with using the index in the user_item matrix
    user_article_idx  = np.where(user_item.loc[user_id] == 1)
    
    #retrieve article ids from article indexes
    user_articles = user_item.loc[user_id].index[user_article_idx].tolist()
    
    #convert to string to keep data types consistent
    article_ids = [str(i) for i in user_articles]
    
    #get article names from article_ids
    article_names = get_article_names(article_ids)
    
    return article_ids, article_names # return the ids and names
```

```python
def user_user_recs(user_id, m=10):
    '''
    INPUT:
    user_id - (int) a user id
    m - (int) the number of recommendations you want for the user
    
    OUTPUT:
    recs - (list) a list of recommendations for the user
    
    Description:
    Loops through the users based on closeness to the input user_id
    For each user - finds articles the user hasn't seen before and provides them as recs
    Does this until m recommendations are found
    
    Notes:
    Users who are the same closeness are chosen arbitrarily as the 'next' user
    
    For the user where the number of recommended articles starts below m 
    and ends exceeding m, the last items are chosen arbitrarily
    
    '''
    # Your code here
    
    #create final list to append recommendation article ids
    recs = []
    
    #retrieve similar users for input user_id
    similar_ids = find_similar_users(user_id)
    
    #article_ids for input user_id
    article_ids_input, article_names_input = get_user_articles(user_id)
    
    #loop through each similar user for input user_id
    for sim_id in similar_ids:
        
        #retrieve similar user's article ids and titles
        article_ids_sim, article_names_sim = get_user_articles(sim_id)
        
        #find articles from similar user not seen by input user_id
        new_recs = np.setdiff1d(article_ids_sim, article_ids_input, assume_unique=True).tolist()
        
        #extend final recommendation list to include these unseen articles
        recs.extend(new_recs)
        
        #break loop if unique list of final recommendations meets the requested number of recommendations
        if len(set(recs)) >= m:
            break
    
    #return "m" number of recommenations in final list
    recs = recs[:m]
    
    
    return recs # return your recommendations for this user_id    
```

```python
def get_top_sorted_users(user_id, df=df, user_item=user_item):
    '''
    INPUT:
    user_id - (int)
    df - (pandas dataframe) df as defined at the top of the notebook 
    user_item - (pandas dataframe) matrix of users by articles: 
            1's when a user has interacted with an article, 0 otherwise
    
            
    OUTPUT:
    neighbors_df - (pandas dataframe) a dataframe with:
                    neighbor_id - is a neighbor user_id
                    similarity - measure of the similarity of each user to the provided user_id
                    num_interactions - the number of articles viewed by the user - if a u
                    
    Other Details - sort the neighbors_df by the similarity and then by number of interactions where 
                    highest of each is higher in the dataframe
     
    '''
    # Your code here
    
    # get row from user_item matrix for the user_id provided
    user_articles = np.array(user_item.loc[user_id])
    
    # compute similarity of each user to the provided user
    dotp = np.dot(user_articles, np.array(user_item).T)
    
    # convert results to Series with an index of user_ids
    simiarlity = pd.Series(dotp, index=user_item.index)
    
    # get the total number of interactions for each user
    interactions_by_user = df.groupby(['user_id'])['article_id'].count()
    
    #create final df with pd.concat with user_ids as the indicies
    final_df = pd.concat([simiarlity,interactions_by_user],axis=1)
    
    #drop input user_id (self) row
    final_df.drop([user_id], inplace = True)
    
    #final column renames
    final_df = final_df.rename(columns={0:'similarity', 'article_id':'num_interactions'})

    # sort by similarity and interactions
    final_df.sort_values(by=['similarity','num_interactions'], ascending=[False,False], inplace=True)
    
    neighbors_df = final_df
    
    return neighbors_df # Return the dataframe specified in the doc_string
```

```python
def user_user_recs_part2(user_id, m=10):
    '''
    INPUT:
    user_id - (int) a user id
    m - (int) the number of recommendations you want for the user
    
    OUTPUT:
    recs - (list) a list of recommendations for the user by article id
    rec_names - (list) a list of recommendations for the user by article title
    
    Description:
    Loops through the users based on closeness to the input user_id
    For each user - finds articles the user hasn't seen before and provides them as recs
    Does this until m recommendations are found
    
    Notes:
    * Choose the users that have the most total article interactions 
    before choosing those with fewer article interactions.

    * Choose articles with the articles with the most total interactions 
    before choosing those with fewer total interactions. 
   
    '''
    # Your code here
    
    #create final list to append recommendation article ids
    recs = []
    
    #retrieve similar users for input user_id
    similar_ids = get_top_sorted_users(user_id)
    
    similar_ids_idx = similar_ids.index.values
    
    #article_ids for input user_id
    article_ids_input, article_names_input = get_user_articles(user_id)
    
    #article ranking table - rank by most interactions
    article_ranked = df.groupby(["article_id"])["user_id"].count().sort_values(ascending=False)
    
    #loop through each similar user for input user_id
    for sim_id in similar_ids_idx:
        
        #retrieve similar user's article ids and titles
        article_ids_sim, article_names_sim = get_user_articles(sim_id)
        
        #find articles from similar user not seen by input user_id
        new_recs = np.setdiff1d(article_ids_sim, article_ids_input, assume_unique=True).tolist()
        
        #convert to float to access article ranking table
        new_recs_float = [float(i) for i in new_recs]
        
        #rank the new recs using the article ranking table - index by article_id list
        new_recs_ranked = article_ranked.loc[new_recs_float].sort_values(ascending=False).index.tolist()
        
        #back to str for use in get_article_names
        new_recs_ranked = [str(i) for i in new_recs_ranked]
        
        #extend final recommendation list to include these unseen articles
        recs.extend(new_recs_ranked)
        
        #break loop if unique list of final recommendations meets the requested number of recommendations
        if len(set(recs)) >= m:
            break
    
    #return "m" number of recommenations in final list
    recs = recs[:m]
    
    rec_names = get_article_names(recs)
    
    #return new_recs_ranked
    return recs, rec_names
```

**Evaluation the Results**

Before looking at additional evaluation methods, we can first evaluate the results of the accuracy scores. One model was calibrated with the entire training data set and the other was with only data from id’s shared by the test train sets. In both instances, we see strong overfitting, as the model with shared ids has approximately a 0.97 accuracy score before decreasing and stabilizing around 0.965, while the model using the entire training dataset starts at 0.991~ and goes to 1. A second indication that we are overfitting is that the data set using the entire training set and testing on itself results in a perfect accuracy score of 1 when approximately 400 latent features are used. With the model using only the shared user-item interactions between test and train, we see that the accuracy actually decrease when we use more latent features, indicating that the addition of two many latent features results in greater overfitting. Ideally, around 200 latent features seems to maximize the results of the models. 

There are a couple of reasons that this overfitting is occurring. The first is that the user-item matrix is heavily imbalanced between the 0 (non-interacted with) and 1 classes (interacted with) as most users have not interacted with a majority of the items. As a result, this training data will lead to a model that will predict a 0 class a majority of the time as the zero class will have a higher probability of being the outcome predicted. The low number of interactions also means that better models will require a much greater sample of users. This leads to the second issue of low sample size. The model using only ids that are shared between the train and test sets highlights this issue of low sample size to an even greater degree. The 20 shared records were not enough to provide a model that was not prone to overfitting as more latent features were added.
